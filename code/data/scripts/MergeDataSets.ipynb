```python
import pandas as pd
```

```python
storm_filtered_data = pd.read_csv('storm_filtered_data.csv')
df_global = pd.read_csv('df_global.csv')
```

```python
storm_filtered_data['ISO_TIME'] = pd.to_datetime(storm_filtered_data['ISO_TIME'])
df_global['dt'] = pd.to_datetime(df_global['dt'])


storm_filtered_data['year_month'] = storm_filtered_data['ISO_TIME'].dt.to_period('M')
df_global['year_month'] = df_global['dt'].dt.to_period('M')


merged_data = pd.merge(
    storm_filtered_data,
    df_global[['year_month', 'LandAverageTemperature', 'LandAndOceanAverageTemperature']],
    on='year_month',
    how='left'
)

merged_data = merged_data.drop(columns=['year_month'])
```

```python
merged_data = merged_data.loc[:, ~merged_data.columns.str.contains('^Unnamed')]
merged_data.head()
```

```python
merged_data.to_csv('merged_data.csv', index=False)
```
